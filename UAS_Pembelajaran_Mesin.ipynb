```python
import snscrape.modules.twitter as sntwitter
import pandas as pd #library untuk membersihkan data mentah dna mengolah data menjadi terstruktur ke dalam tabel
```

```python
pd.options.display.max_colwidth = 500 #pandas men-setting agar maxsimal kolom yang ditampilkan hanya sampai 500
```

```python
query = "(from:collegemenfess) until:2022-12-11 since:2022-12-05" #query diambil dari twitter collegemenfess sejak tanggal 5 desember hingga 11 desember
tweets = []
limit = 5000 #batas yang diambil yaitu 5000 data

%time
try:
    print("Start crawling") #memulai crawling data/semacam pengambilan data dari media sosial kemudian di kumpulkan untuk diolah pada proses selanjutnya
    for tweet in sntwitter.TwitterSearchScraper(query=query).get_items(): #pada sntwitter tadi menghubungkan dengan query untuk mendapatkan item yang diperlukan
        if len(tweets) == limit: #jika panjang jumlah tweets sama dengan limit 5000 data
            break #berhenti
        else:
            tweets.append([tweet.date, tweet.user.username, tweet.content])
    df = pd.DataFrame(tweets, columns=['datetime', 'username', 'content']) # membuatkan tabel 2 dimensi dengan isian datetime, username, content
except Exception as e: #untuk menangani kesalahan tak terduga dalam program Python
    print(e) #tampilkan exception
print("Finished") #tampilkan tulisan finished
```

```python
import os #mengakses banyak fungsi sistem operasi.
import re #memeriksa apakah suatu string berisikan pola pencarian yang spesifik
from textblob import TextBlob #melakukan berbagai proses terhadap data teks mulai dari yang sederhana seperti tokenisasi (pemotongan kata) sampai analisa sentiment.
import matplotlib.pyplot as plt #membuat visualisasi data berbentuk 2D/3D
from wordcloud import WordCloud #ukuran kata yang akan muncul, besar kecil kata dalam wordcloud akan merepresentasi-kan jumlah kemunculan kata pada suatu teks
import seaborn as sns #menciptakan visualisasi data kualitas tinggi
from sklearn.preprocessing import LabelEncoder #Pengkodean label mengacu pada proses mengubah label kata menjadi bentuk numerik.
```

```python
#CASE FOLDING :Pengubahan Huruf besar ke huruf kecil 
def clean_lower(lwr):
    lwr = lwr.lower() # mengubah menjadi lowercase text
    return lwr
```

```python
#Tokenizing : penghapusan karakter (@, '', "", dll)
#Remove Puncutuation, penghapusan tanda baca
clean_spcl = re.compile('[/(){}\[\]\|@,;][c][m]')
clean_symbol = re.compile('[^0-9a-z]')
def clean_punct(text):
    text = clean_spcl.sub('', text)
    text = clean_symbol.sub(' ', text)
    return text
```

```python
#Tokenizing : Penghapusan Spasi yang berlebihan 
def _normalize_whitespace(text):
    corrected = str(text)
    corrected = re.sub(r"//t",r"\t", corrected) #penghapusan tab
    corrected = re.sub(r"( )\1+",r"\1", corrected)
    corrected = re.sub(r"(\n)\1+",r"\1", corrected)
    corrected = re.sub(r"(\r)\1+",r"\1", corrected)
    corrected = re.sub(r"(\t)\1+",r"\1", corrected)
    return corrected.strip(" ")
```

```python
#Tokenizing : Penghapusan number
def remove_number(text):
    return re.sub(r"\d", "", text)
```

```python
#Membuat kolom baru untuk hasil konten yang telah dilakukan preprocesing
df['hasil_tweet'] = df['content'].apply(clean_lower).apply(clean_punct).apply(_normalize_whitespace).apply(remove_number)

df.head()
```

```python
# Membuat nilai dari setiap komentar / tweet dengan nilai dari nilai subjektif dan polarity

# Fungsi untuk Membuat nilai subjektif //bertujuan untuk mengukur apakah kalimat tersebut fakta atau opini 
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

# Fungsi untuk membuat nilai polarity //bertujuan untuk mengukur data perasaan/sentiment 
def getPolarity(text):
    return TextBlob(text).sentiment.polarity

# Mmebuat 2 Kolom baru untuk menyimpan nilai subjektif dan polarity
df['Subjectivity'] = df['hasil_tweet'].apply(getSubjectivity)
df['Polarity'] = df['hasil_tweet'].apply(getPolarity)

# Cek Data
df.head(30)
```

```python
# Membuat fungsi untuk memberi label sentimen dari data tweet
def getSentiment(score):
    if score < 0 :
        return 'Negative'
    elif score == 0 :
        return 'Netral'
    else :
        return 'Positive'
# Aplly fungsi ke dalam data 
df['Sentiment'] = df['Polarity'].apply(getSentiment)
# Cek data
df.head(30)
```

## **LABELING**

```python
# Memilih Data mana saja yang akan digunakan
df = pd.DataFrame({

    'hasil_tweet' : df['hasil_tweet'],
    'Sentiment' : df['Sentiment']
})
df.head()
```

```python
#Mengubah dara kategorik sentiment menjadi data numerik 
encode = LabelEncoder()
content = encode.fit_transform(df['Sentiment'])
df['Label_Sentiment']  = pd.DataFrame(content)
df.head()
```

```python
# #Memisahkan Fitur dengan Label
X = df['hasil_tweet'].values
y = df['Label_Sentiment'].values
```

```python
#EKSTRASI FITUR
#Ekstrasi Fitur yang kami dengan menggunakan fungsi TfidfVectorizer 
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# Split data training dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)

# Inisiasi TfidfVectorizer
bow = TfidfVectorizer(stop_words='english')

# Fitting dan transform X_train dengan TfidfVectorizer
X_train = bow.fit_transform(X_train)

# Transform X_test
X_test = bow.transform(X_test)
```

```python
#Klasifikasi yang digunakan adalah algoritma Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

#Fit ke model
rdf_model = RandomForestClassifier().fit(X_train, y_train)

# Prediksi dengan data training
y_pred_train = rdf_model.predict(X_train)

# Evaluasi akurasi data training
acc_train = accuracy_score(y_train, y_pred_train)

# Prediksi dengan data testing
y_pred_test = rdf_model.predict(X_test)

# Evaluasi akurasi data testing
acc_test = accuracy_score(y_test, y_pred_test)

# Print hasil evaluasi
print(f'Hasil akurasi data train: {acc_train}')
print(f'Hasil akurasi data test: {acc_test}')
```

```python
#Melakukan Evaluasi dengan confusion matrix
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_test))
```
